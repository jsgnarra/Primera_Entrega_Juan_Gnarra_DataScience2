```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Agrupar por subcategoría
rentabilidad_por_producto = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)

# Tomar el top 10 de productos más rentables
top_productos = rentabilidad_por_producto.head(10)

# Crear la paleta de colores del más rentable (verde) al menos rentable (rojo)
# Usamos 'RdYlGn' invertido (de verde a rojo)
colores = sns.color_palette("RdYlGn_r", n_colors=len(top_productos))

# Gráfico de torta
plt.figure(figsize=(8, 8))
plt.pie(top_productos, labels=top_productos.index, autopct='%1.1f%%',
        startangle=90, colors=colores)
plt.title('Participación de rentabilidad por subcategoría (Top 10)')
plt.axis('equal')
plt.show()

```

```python
# Target
y = df['Quantity']

# Features seleccionadas
X = df[['Amount', 'Profit', 'Category', 'Sub-Category', 'PaymentMode', 'State', 'City']]

# One-hot encoding
X_encoded = pd.get_dummies(X, drop_first=True)

# División de datos en entrenamiento y prueba
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Dividir en training y test (80/20)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# Predicciones
y_pred = modelo.predict(X_test)

# Evaluación del modelo
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("📊 Evaluación del modelo:")
print(f"MAE (Error absoluto medio): {mae:.2f}")
print(f"RMSE (Raíz del error cuadrático medio): {rmse:.2f}")
print(f"R² (Coeficiente de determinación): {r2:.2f}")
```

```python
# ---------------------------------------------------------
# 🔁 Modelo 2: Random Forest Regressor + Validación cruzada
# ---------------------------------------------------------

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score, GridSearchCV

# Instanciar modelo base
rf = RandomForestRegressor(random_state=42)

# Validación cruzada (con 5 folds) para ver performance general
cv_scores = cross_val_score(rf, X_encoded, y, cv=5, scoring='r2')
print(f"R² promedio en CV (Random Forest base): {cv_scores.mean():.2f}")

# ---------------------------------------
# 🔧 Optimización con GridSearchCV
# ---------------------------------------

param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}

grid = GridSearchCV(estimator=rf, param_grid=param_grid,
                    cv=3, scoring='r2', n_jobs=-1, verbose=1)

# Entrenar con datos de entrenamiento
grid.fit(X_train, y_train)

# Mejor modelo encontrado
mejor_rf = grid.best_estimator_

# Predicción en test
y_pred_rf = mejor_rf.predict(X_test)

# Métricas
mae_rf = mean_absolute_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
r2_rf = r2_score(y_test, y_pred_rf)

print("\n📊 Evaluación del modelo: Random Forest Optimizado")
print(f"MAE: {mae_rf:.2f}")
print(f"RMSE: {rmse_rf:.2f}")
print(f"R²: {r2_rf:.2f}")
```

```python
# Comparar ambos modelos
comparacion_modelos = pd.DataFrame({
    'Modelo': ['Regresión Lineal', 'Random Forest Optimizado'],
    'MAE': [mae, mae_rf],
    'RMSE': [rmse, rmse_rf],
    'R²': [r2, r2_rf]
})

comparacion_modelos
```

```python
# Calcular promedios reales de Amount y Profit por categoría
promedios_por_categoria = df.groupby('Category')[['Amount', 'Profit']].mean().reset_index()

# Crear DataFrame de predicción con variables más realistas
X_cat_realista = pd.DataFrame({
    'Amount': promedios_por_categoria['Amount'],
    'Profit': promedios_por_categoria['Profit'],
    'Category': promedios_por_categoria['Category'],
    'Sub-Category': ['Chairs'] * len(promedios_por_categoria),  # Fijo
    'PaymentMode': ['Cash'] * len(promedios_por_categoria),     # Fijo
    'State': ['California'] * len(promedios_por_categoria),
    'City': ['Los Angeles'] * len(promedios_por_categoria),
})

# One-hot encoding
X_cat_realista_encoded = pd.get_dummies(X_cat_realista)
X_cat_realista_encoded = X_cat_realista_encoded.reindex(columns=X_encoded.columns, fill_value=0)

# Predicciones con valores más representativos por categoría
ventas_categoria_real = mejor_rf.predict(X_cat_realista_encoded)

# DataFrame para gráfico
df_pred_cat_real = pd.DataFrame({
    'Categoría': promedios_por_categoria['Category'],
    'Ventas Predichas': ventas_categoria_real
})

# Gráfico
plt.figure(figsize=(8, 5))
sns.barplot(data=df_pred_cat_real, x='Categoría', y='Ventas Predichas', palette='Set1')
plt.title('🔮 Ventas esperadas por categoría para el próximo año (simulación realista)')
plt.ylabel('Unidades estimadas')
plt.xlabel('Categoría')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# ✅ Asegurar formato de fecha y extraer columna 'Año'
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Año'] = df['Order Date'].dt.year

# ✅ Filtrar registros del año 2024
df_2024 = df[df['Año'] == 2024].copy()

# ✅ Eliminar columna target real
df_2024.drop(columns=['Quantity'], inplace=True)

# ✅ Preparar features igual que en entrenamiento
X_2025 = df_2024[['Amount', 'Profit', 'Category', 'Sub-Category', 'PaymentMode', 'State', 'City']]
X_2025_encoded = pd.get_dummies(X_2025)
X_2025_encoded = X_2025_encoded.reindex(columns=X_encoded.columns, fill_value=0)

# ✅ Predecir cantidad de ventas por orden
ventas_pred_2025 = mejor_rf.predict(X_2025_encoded)

# ✅ Sumar predicciones para obtener total estimado 2025
ventas_totales_2025_corr = ventas_pred_2025.sum()

print(f"🔮 Ventas estimadas por el modelo para 2025 (simulación más realista): {int(ventas_totales_2025_corr)} unidades")
```

```python
# 📦 Ventas reales por año
ventas_por_anio = df.groupby('Año')['Quantity'].sum().reset_index()
ventas_por_anio.rename(columns={'Quantity': 'Ventas'}, inplace=True)

# 💰 Profit real por año
profit_por_anio = df.groupby('Año')['Profit'].sum().reset_index()
profit_por_anio.rename(columns={'Profit': 'Ganancia'}, inplace=True)

# 🔮 Agregar estimación 2025 (usando la simulación realista)
ventas_2025 = 20996  # predicción del modelo
profit_unitario = df['Profit'].sum() / df['Quantity'].sum()
ganancia_2025 = ventas_2025 * profit_unitario

ventas_por_anio = pd.concat([ventas_por_anio, pd.DataFrame({'Año': [2025], 'Ventas': [ventas_2025]})], ignore_index=True)
profit_por_anio = pd.concat([profit_por_anio, pd.DataFrame({'Año': [2025], 'Ganancia': [ganancia_2025]})], ignore_index=True)

# 🧩 Combinar todo
df_doble = ventas_por_anio.merge(profit_por_anio, on='Año')
```

## 📊 Gráfico final con doble eje: Ventas y Ganancias

```python
# 📊 Gráfico con doble eje Y: Ventas (azul) y Ganancia (rojo)
fig, ax1 = plt.subplots(figsize=(10, 6))

# Eje izquierdo: Ventas (azul)
color_ventas = 'blue'
ax1.set_xlabel('Año')
ax1.set_ylabel('Ventas (unidades)', color=color_ventas)
ax1.plot(df_doble['Año'], df_doble['Ventas'], color=color_ventas, marker='o', label='Ventas')
ax1.tick_params(axis='y', labelcolor=color_ventas)

# Eje derecho: Ganancia (rojo)
ax2 = ax1.twinx()
color_profit = 'red'
ax2.set_ylabel('Ganancia (total $)', color=color_profit)
ax2.plot(df_doble['Año'], df_doble['Ganancia'], color=color_profit, marker='o', label='Ganancia')
ax2.tick_params(axis='y', labelcolor=color_profit)

# Título
plt.title('📊 Evolución de Ventas y Ganancias Totales (2020–2025)')
fig.tight_layout()
plt.grid(True)
plt.show()
```
